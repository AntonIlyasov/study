## Задание 4. Представление предложений в виде векторов


Дано предложение: <code>the quick brown fox jumps over the lazy dog</code>. Вам нужно представить это предложение в виде вектора. Для этого найдите вектор каждого слова в модели, а затем усредните векторы покомпонентно.


````python
import numpy as np

from gensim.models import Word2Vec
from nltk.tokenize import word_tokenize

# Предложение
sentence = "chain is only as strong as its weakest link"

# Токенизация предложения
tokens = word_tokenize(sentence)

# Получение векторного представления каждого слова
word_vectors = [w[token] for token in tokens]

# Усреднение векторов по компонентам
sentence_vector = np.mean(word_vectors, axis=0)

# print(sentence_vector)

# Предложение
sentence2 = "Actions speak louder than words"

# Токенизация предложения
tokens2 = word_tokenize(sentence2)

# Получение векторного представления каждого слова
word_vectors2 = [w[token] for token in tokens2]

# Усреднение векторов по компонентам
sentence_vector2 = np.mean(word_vectors2, axis=0)

# print(sentence_vector2)

from scipy.spatial.distance import cosine

# Вычисление косинусного расстояния между векторами
cosine_distance = cosine(sentence_vector, sentence_vector2)

print("Косинусное расстояние между вектором 'sentence_vector' и вектором предложения sentence_vector2:", cosine_distance)
````

Посчитаем корреляцию оценок сходства в модели google-news-vectors с оценками аннотаторов в датасете, а затем - корреляцию сходства в модели на основе британского национального корпуса с оценками аннотаторов в датасете. (используем только те слова из wordsim, для которых находятся векторы на британском корпусе, помеченные как существительные!)

````python
import numpy as np
from scipy.stats import spearmanr
from gensim.models import KeyedVectors

# Load word vector models
british_model = KeyedVectors.load_word2vec_format("model.bin", binary=True)  # Путь к модели британского корпуса
other_model = KeyedVectors.load_word2vec_format("GoogleNews-vectors-negative300.bin", binary=True)  # Путь к другой модели, например, gpt-3.5-turbo

# Read word pairs from wordsim_similarity_goldstandard.txt
word_pairs = []
with open("wordsim_similarity_goldstandard.txt", "r") as file:
    for line in file:
        word1, word2, score = line.strip().split("\t")
        word_pairs.append((word1.lower(), word2.lower(), float(score)))

# Extract subset of word pairs from indices 14 to 113
word_pairs_subset = word_pairs[14:114]

# Filter word pairs based on vector existence and part-of-speech
filtered_word_pairs = []
for word1, word2, _ in word_pairs_subset:
    if word1 in british_model and word2 in british_model:
        if 'n' in british_model.get_vector(word1).flags and 'n' in british_model.get_vector(word2).flags:
            filtered_word_pairs.append((word1, word2))

# Calculate cosine similarity for filtered word pairs in both models
model_similarities_british = []
model_similarities_other = []
for word1, word2 in filtered_word_pairs:
    similarity_british = british_model.similarity(word1, word2)
    similarity_other = other_model.similarity(word1, word2)
    model_similarities_british.append(similarity_british)
    model_similarities_other.append(similarity_other)

# # Compute Spearman correlation between model similarity scores and human annotations
# human_scores = [score for _, _, score in word_pairs_subset]
# spearman_corr, _ = spearmanr(human_scores, model_similarities_other)

# print("Spearman correlation between model similarity scores and human annotations:", spearman_corr)
````
